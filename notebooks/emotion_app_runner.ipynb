```python
import os
import shutil
from pyngrok import ngrok
from pyngrok.conf import PyngrokConfig
import subprocess # Import subprocess for explicit command execution

# IMPORTANT: Replace <YOUR_AUTHTOKEN> with your actual ngrok authentication token
# You can find your authtoken on your ngrok dashboard after signing up.
NGROK_AUTH_TOKEN = "350xepgfaos6TeMXjLyYWb0UeUX_4PhBYeDayD7dBBY5KkTGy"

# Define a temporary directory for ngrok and its configuration
NGROK_TEMP_DIR = "/tmp/ngrok_colab_config"

# Ensure a clean state by removing the temporary directory if it exists
if os.path.exists(NGROK_TEMP_DIR):
    print(f"Cleaning up existing ngrok temporary directory: {NGROK_TEMP_DIR}")
    shutil.rmtree(NGROK_TEMP_DIR)
os.makedirs(NGROK_TEMP_DIR, exist_ok=True)

# Create a PyngrokConfig object to specify custom paths
pyngrok_config = PyngrokConfig()
pyngrok_config.ngrok_path = os.path.join(NGROK_TEMP_DIR, "ngrok")
pyngrok_config.config_path = os.path.join(NGROK_TEMP_DIR, "ngrok.yml")

# Ensure ngrok is installed in the specified temporary directory using pyngrok
print(f"Installing ngrok to {pyngrok_config.ngrok_path} using pyngrok...")
ngrok.install_ngrok(pyngrok_config)
print("ngrok installation complete.")

# --- Explicitly set the auth token using subprocess to diagnose further ---
print("Attempting to set ngrok authentication token directly...")
try:
    # Construct the command to set the authtoken using the specific ngrok binary and config path
    auth_command = [
        pyngrok_config.ngrok_path,
        'config',
        'add-authtoken',
        NGROK_AUTH_TOKEN,
        '--log',
        'stdout',
        '--config',
        pyngrok_config.config_path
    ]

    # Execute the command and capture output for debugging
    result = subprocess.run(auth_command, capture_output=True, text=True, check=True)
    print("ngrok authtoken command output:")
    print(result.stdout)
    if result.stderr:
        print("ngrok authtoken command stderr:")
        print(result.stderr)
    print("Authentication token set via subprocess.")
except subprocess.CalledProcessError as e:
    print(f"Error setting ngrok authtoken directly: {e.cmd}")
    print(f"Stdout: {e.stdout}")
    print(f"Stderr: {e.stderr}")
    print("This indicates a problem with the ngrok binary or its configuration command.")
    raise # Re-raise the exception to halt execution and report the error
# --- End explicit auth token setting ---

# Connect ngrok using the custom PyngrokConfig
print("Connecting ngrok tunnel...")

# Kill any existing ngrok tunnels to free up endpoints if hit the limit
print("Attempting to kill any existing ngrok processes...")
ngrok.kill()
print("Existing ngrok processes killed (if any).")

public_url = ngrok.connect(8501, pyngrok_config=pyngrok_config)
print(f"ngrok tunnel established at: {public_url}")

print(public_url)
```
